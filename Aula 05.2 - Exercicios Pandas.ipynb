```python
import pandas as pd
import openpyxl
```

**1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)**

```python
df = pd.read_csv('dados/alunos3.csv', sep=';', decimal=',')
df
```

```python
df['Media_provas'] = (df['Prova_1'] + df['Prova_2'] + df['Prova_3'] + df['Prova_4'])/4
df
```

**2) Quem foram os alunos que obtiveram a maior e a menor média**

```python
maiorMedia = df["Media_provas"] == df['Media_provas'].max()
menorMedia = df["Media_provas"] == df['Media_provas'].min()
print(df[maiorMedia])
print(df[menorMedia])
```

**3) Agora una este dataframe com o `cadastro_alunos.xlsx`**

```python
df2= pd.read_excel('dados/cadastro_alunos.xlsx')

df3 = pd.merge(df2, df, on="RA",how="left") 
df3
```

**4) Qual a média entre as Media_provas dentro do público feminino? e masculino?**

```python
feminino = df3[df3['Sexo'] == 'Feminino']['Media_provas'].mean()
masuclino = df3[df3['Sexo'] == 'Masculino']['Media_provas'].mean()

print(f'Média feminina: {feminino}')
print(f'Média masculina: {masuclino}')
```

**5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?**

```python
mediaIdade = df3[df3['Media_provas'] >= 7]['Idade'].mean()
mediaIdade
```

**6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?**

```python
indexMaiorMedia = (df3["Media_provas"] == df3['Media_provas'].max()) # Inspirado na questão 2; índice da maior nota
cidadeComMaiorMedia = df3[indexMaiorMedia]['Cidade']
maiorMedia = df3[indexMaiorMedia]["Media_provas"]

print(f'A maior média foi de {maiorMedia}')
print(f'A cidade foi {cidadeComMaiorMedia}')
```
